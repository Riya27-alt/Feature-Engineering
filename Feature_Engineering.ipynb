Feature Engineering.
1. What is a parameter?
A parameter is a value that the model learns from the training data (like slope and intercept in a linear regression model).

2. What is correlation?
Correlation is a statistical measure that shows how strongly two variables are related. It ranges from -1 to +1.

3. What does negative correlation mean?
Negative correlation means as one variable increases, the other decreases (e.g., more exercise, less weight).

4. Define Machine Learning. What are the main components?
Machine Learning is a method where machines learn from data to make predictions.
Main components:

Data

Model/Algorithm

Loss Function

Optimizer

Training & Testing Process

5. How does loss value help determine model quality?
A low loss value means the model predictions are close to the actual values, indicating a good model.

6. What are continuous and categorical variables?
Continuous: Numeric values (e.g., height, temperature).

Categorical: Categories or labels (e.g., gender, country).

7. How do we handle categorical variables?
Common techniques:

Label Encoding

One-Hot Encoding

Ordinal Encoding

8. What is training and testing a dataset?
Training set: Used to teach the model.

Testing set: Used to evaluate model performance.

9. What is sklearn.preprocessing?
A module in scikit-learn used for preprocessing data (e.g., scaling, encoding).

10. What is a test set?
A test set is a part of the dataset used to evaluate the trained model’s performance.

15. What is causation? Correlation vs. Causation?
Causation: One variable causes the other.

Correlation: They change together but one doesn't cause the other.

Example: Ice cream sales and drowning are correlated (both increase in summer) but one doesn’t cause the other.

20. What is feature scaling?
It standardizes or normalizes data.
Helps algorithms perform better, especially those using distance (e.g., SVM, KNN).
